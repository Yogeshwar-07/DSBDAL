```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score, confusion_matrix, accuracy_score, recall_score
```

```python
df = pd.read_csv("Social_Media_Advertising.csv")
```

```python
df = pd.read_csv("Social_Media_Advertising.csv", nrows=1000)
```

```python
if df.select_dtypes(include=['object']).shape[1] > 0:
    df = pd.get_dummies(df, drop_first=True)
```

```python
if df.isnull().sum().sum() > 0:
    df = df.dropna()
```

```python
cov_matrix = df.cov()
print("Covariance Matrix:\n", cov_matrix)
```

```python
df.columns
```

```python
X = df.drop(columns=["Impressions"])                 # Replace "Purchased" with target column name
Y = df["Impressions"]
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
xtrain = scaler.fit_transform(xtrain)
xtest = scaler.transform(xtest)
```

```python
logreg = LogisticRegression()
# fit the model with data
logreg.fit(xtrain,ytrain)
# y_pred=logreg.predict(xtest)
```

```python
# Predict on training data
y_pred_train = logreg.predict(xtrain)

# Predict on test data
y_pred_test = logreg.predict(xtest)
```

```python
# Precision
precision_train = precision_score(ytrain, y_pred_train, average='weighted')
precision_test = precision_score(ytest, y_pred_test, average='weighted')

# Recall
recall_train = recall_score(ytrain, y_pred_train, average='weighted')
recall_test = recall_score(ytest, y_pred_test, average='weighted')
```

```python
from sklearn.metrics import confusion_matrix

# Confusion Matrix for Training and Testing
cm_train = confusion_matrix(ytrain, y_pred_train)
cm_test = confusion_matrix(ytest, y_pred_test)

# Now print
print("Confusion Matrix (Train):\n", cm_train)
print("Confusion Matrix (Test):\n", cm_test)
```

```python
from sklearn.metrics import accuracy_score

# Ensure Predictions Exist
y_pred_train = logreg.predict(xtrain)
y_pred_test = logreg.predict(xtest)

# Compute Accuracy
accuracy_train = accuracy_score(ytrain, y_pred_train)
accuracy_test = accuracy_score(ytest, y_pred_test)

# Now print
print("\nAccuracy (Train):", accuracy_train)
print("Accuracy (Test):", accuracy_test)
```
